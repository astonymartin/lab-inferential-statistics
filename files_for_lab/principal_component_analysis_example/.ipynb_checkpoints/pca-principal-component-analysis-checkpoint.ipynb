Before we apply PCA we need to standardize/scale the data. For this we can use sklearn's preprocessing library 


```python
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(numerics)
scaled = transformer.transform(numerics)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca = PCA()
pca.fit(scaled)
```

```python
# variations = pca.explained_variance_
variations = pca.explained_variance_ratio_*100
variations = np.round(variations, decimals=1)
variations
```

```python
presion = np.random.normal(loc=130.1, scale=21.21, size=100)
```

```python
presion_samples = []

for i in range(100):
    sample = np.random.choice(presion, 100)
    sample_mean = sum(sample) / 100
    presion_samples.append(sample_mean)
```

```python
plt.hist(presion_samples)
```

```python
np.mean(presion_samples)
```

```python
presion_samples = pd.DataFrame(presion_samples)
```

```python
presion_samples
```

Ho: mu = 120

```python
mu = 120
```

```python
alpha = 0.05
```

```python
mean = presion_samples.mean()
s = presion_samples.std()
n = presion_samples.shape[0]
```

```python
mean
```

```python
s
```

```python
n
```

```python
stat = (mean - mu) / (s/ np.sqrt(n))
stat 
```

```python
p_value = st.t.sf(abs(stat), n-1)*2
p_value
```

```python
if p_value > alpha:
    print("We are not able to reject the null hypothesis")
else:
    print("We reject the null hypotesis")
```
